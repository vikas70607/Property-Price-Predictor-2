```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from turtle import color
from sklearn import linear_model
```

### Loading data and Preprocessing

```python
data = pd.read_csv("data.csv")
df = data.dropna()
df.head()
```

```python
x = df[['Area','Bathroom']]
y = df['Price']
```

### Plotting 
Plotting the points with Area and Bathroom Independent variables and Price as a dependent variable.

```python
ax = plt.axes(projection ='3d')
ax.scatter(x['Area'],x['Bathroom'],y)
ax.set(xlabel= 'Area', ylabel= 'Bathroom', zlabel= 'Price')
```

### Plotting the Regression plane

```python



regr = linear_model.LinearRegression()
regr.fit(x.values, y.values)
cf = regr.coef_
b = regr.intercept_

ax = plt.axes(projection ='3d')
ax.scatter(x['Area'],x['Bathroom'],y)
ax.plot_trisurf(x['Area'],x['Bathroom'], ((x['Area']*cf[0]) + (x['Bathroom']*cf[1]) + b),color =  'red')
plt.show()
```

### Prediction

```python
#Price of the flat with 2 Bathroom and 750 sq.feet area 
Price_1 = regr.predict([[750,2]])
print(Price_1)
#Price of the flat with 3 Bathroom and 1000 sq.feet area 
Price_2 = regr.predict([[1000,3]])
print(Price_1)
```

### Accuracy 
Outputs the **R<sup>2</sup>** score

```python
score = regr.score(x.values,y.values)
print(score)
```
